```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
import math

```

```python
def fft_with_show(image_path):
    img = cv2.imread(image_path)
    imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)
    magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))
    
    plt.figure(figsize=(12,12))
    plt.subplot(141)
    plt.imshow(imgrgb)
    plt.title("input")
    
    plt.subplot(142)
    plt.imshow(magnitude_spectrum)
    plt.title("magnitude")
    
    plt.subplot(143)
    plt.imshow(dft_shift[:,:,0])
    plt.title("dft_shift")
    
    plt.subplot(144)
    plt.imshow(dft[:,:,0])
    plt.title("dft")
    plt.show()
    
    return magnitude_spectrum, dft_shift, dft
```

```python
def fft_resove(img):
    
    dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)
    magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))

    rows, cols = img.shape
    crow,ccol = int(rows/2) , int(cols/2)
    
    mask = np.zeros((rows,cols,2),np.uint8)
    mask[crow-30:crow+30, ccol-30:ccol+30] = 1

    fshift = dft_shift*mask
    f_ishift = np.fft.ifftshift(fshift)

    img_back = cv2.idft(f_ishift)
    img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])
    img_back = cv2.dft(np.float32(img_back), flags=cv2.DFT_COMPLEX_OUTPUT)
    low_dft_shift = np.fft.fftshift(img_back)
    #low_dft_shift += 1
    low_magnitude_spectrum = 20 * np.log(cv2.magnitude(low_dft_shift[:, :, 0], low_dft_shift[:, :, 1]))
    
    
    mask2 = np.ones((rows,cols,2),np.uint8)
    mask2[crow-30:crow+30, ccol-30:ccol+30] = 0
    fshift = dft_shift*mask2
    f_ishift = np.fft.ifftshift(fshift)

    img_back = cv2.idft(f_ishift)
    img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])
    img_back = cv2.dft(np.float32(img_back), flags=cv2.DFT_COMPLEX_OUTPUT)
    high_dft_shift = np.fft.fftshift(img_back)
    #high_dft_shift += 1
    high_magnitude_spectrum1 = 20 * np.log(cv2.magnitude(high_dft_shift[:, :, 0], high_dft_shift[:, :, 1]))

    return magnitude_spectrum, low_magnitude_spectrum, high_magnitude_spectrum1
```

```python
from pywt import dwt2, idwt2
```

```python
def dwt(image, method = "haar"):
    img = cv2.imread(image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cA,(cH,cV,cD)=dwt2(img, method)
    
    plt.figure(figsize=(12,12))
    plt.subplot(151)
    # 小波变换之后，低频分量对应的图像：
    plt.imshow(np.uint8(cA/np.max(cA)*255))
    # 小波变换之后，水平方向高频分量对应的图像：
    plt.subplot(152)
    plt.imshow(np.uint8(cH/np.max(cH)*255))
    #cv2.imwrite('lena_h.png',)
    # 小波变换之后，垂直平方向高频分量对应的图像：
    plt.subplot(153)
    plt.imshow(np.uint8(cV/np.max(cV)*255))
    #cv2.imwrite('lena_v.png',np.uint8(cV/np.max(cV)*255))
    # 小波变换之后，对角线方向高频分量对应的图像：
    plt.subplot(154)
    plt.imshow(np.uint8(cD/np.max(cD)*255))
    
    
    # 根据小波系数重构回去的图像
    rimg = idwt2((cA,(cH,cV,cD)), 'haar')
    plt.subplot(155)
    plt.imshow(np.uint8(rimg))
    plt.show()
    return cA,(cH,cV,cD)
```

```python
def fft_resove(image_path):
    img = cv2.imread(image_path)
    imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)
    magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))

    rows, cols = img.shape
    crow,ccol = int(rows/2) , int(cols/2)
    
    mask = np.zeros((rows,cols,2),np.uint8)
    mask[crow-30:crow+30, ccol-30:ccol+30] = 1
    fshift = dft_shift*mask
    f_ishift = np.fft.ifftshift(fshift)

    img_back = cv2.idft(f_ishift)
    img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])
    img_back = cv2.dft(np.float32(img_back), flags=cv2.DFT_COMPLEX_OUTPUT)
    low_dft_shift = np.fft.fftshift(img_back)
    
    low_magnitude_spectrum = 20 * np.log(cv2.magnitude(low_dft_shift[:, :, 0], low_dft_shift[:, :, 1]))
    
    
    mask2 = np.ones((rows,cols,2),np.uint8)
    mask2[crow-30:crow+30, ccol-30:ccol+30] = 0
    fshift = dft_shift*mask2
    f_ishift = np.fft.ifftshift(fshift)
    img_back = cv2.idft(f_ishift)
    img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])
    
    dft1 = cv2.dft(np.float32(img_back), flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_shift1 = np.fft.fftshift(dft1)
    
    heigh_magnitude_spectrum1 = 20 * np.log(cv2.magnitude(dft_shift1[:, :, 0], dft_shift1[:, :, 1]))
    return magnitude_spectrum, low_magnitude_spectrum, heigh_magnitude_spectrum1
```
